```python
#Encoding  gender,country,residence and prevEducation
print(df['gender'].unique())
print(df['country'].unique())
print(df['residence'].unique())
print(df['prevEducation'].unique())
```

```python
#standardize the values at gender ,residency and prevEducation
#Replace all variations of  gender with 'male'
df['gender'] = df['gender'].str.lower().replace({'m': 'male'})
print(df['gender'].unique())

#Replace all variations of residence with 'BIResidence'
df['residence'] = df['residence'].replace({'BI Residence': 'BIResidence',
    'BI-Residence': 'BIResidence',
    'BI_Residence': 'BIResidence',})
print(df['residence'].unique())

#replace varations of prevEducation
df['prevEducation'] = df['prevEducation'].replace({'High School':'HighSchool',
      'DIPLOMA':'Diploma','diploma':'Diploma','Diplomaaa':'Diploma','Bachelors':
                                                  'Barrrchelors' })
print(df['prevEducation'].unique())
# df['country'].unique()
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
```

```python
df['gender']=le.fit_transform(df['gender'])
print(df['gender'].unique())
df['residence']=le.fit_transform(df['residence'])
print(df['residence'].unique())
df['prevEducation']=le.fit_transform(df['prevEducation'])
print(df['prevEducation'].unique())
df['country']=le.fit_transform(df['country'])
print(df['country'].unique())
```

```python
#Copy
df1 =df.copy()
df2 =df.copy()
df3 =df.copy()
df4 =df.copy()
df5 =df.copy()
```

```python
#Split X  value (independent values)
x = df.drop('gender', axis=1)
x
```

```python
#split Y value(target value)
y = df['gender']
y
```

```python
#train modules
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
M1_knn = KNeighborsClassifier(n_neighbors=7)
M2_svc = SVC(kernel='linear',C=2)
M3_LR = LogisticRegression()
```

```python
M1_knn.fit(x_train, y_train)
```

```python
M2_svc.fit(x_train, y_train)
```

```python
M3_LR.fit(x_train,y_train)
```

```python
y_predict_knn = M1_knn.predict(x_test)
y_predict_svc = M2_svc.predict(x_test)
y_predict_LR = M3_LR.predict(x_test)
```

```python
print("KNN Predict:",y_predict_knn)
print("SVC Predict:",y_predict_svc)
print("LR predict:",y_predict_LR)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
cm_knn = confusion_matrix(y_test,y_predict_knn)
cm_svc = confusion_matrix(y_test,y_predict_svc)
cm_LR = confusion_matrix(y_test,y_predict_LR)
print("confusion matrics for KNN",cm_knn)

print("confusion matrics for svc",cm_svc)
print("confusion matrics for LR",cm_LR)
```

```python
acs_knn =accuracy_score(y_test,y_predict_knn)
acs_svc =accuracy_score(y_test,y_predict_svc)
acs_lr =accuracy_score(y_test,y_predict_LR)

print("accuracy score KNN:",acs_knn)
print("accuracy score SVM:",acs_svc)
print("accuracy score LR:",acs_lr)
```

```python
# Best model   :  SVM as it get 80%
# Average model:  Logistic Regression as it get 73%
# Weakest model:  KNN as it get 66%
```

```python
import matplotlib.pyplot as plt

# Accuracy scores
accuracy_scores = [acs_knn, acs_svc, acs_lr]
model_names = ['KNN', 'SVM', 'Logistic Regression']

# Create bar chart
plt.figure(figsize=(8, 5))
bars = plt.bar(model_names, accuracy_scores, color=['skyblue', 'lightgreen', 'salmon'])

# Add value labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.01, f"{yval:.2f}", ha='center', fontsize=12)

# Add chart title and labels
plt.title('Accuracy Scores of Different Models', fontsize=14)
plt.ylabel('Accuracy', fontsize=12)
plt.ylim(0, 1.1)  # to leave space above the bars for labels

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
